```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Define the tickers and the start date
sp500_ticker = "^GSPC"  # Ticker for S&P 500
gsci_ticker = "^SPGSCI"     # An ETF approximating GSCI
start_date = "1991-05-01"

# Fetch the historical data from Yahoo Finance
sp500_data = yf.download(sp500_ticker, start=start_date)['Adj Close']
gsci_data = yf.download(gsci_ticker, start=start_date)['Adj Close']

# Calculate daily returns in percentage
sp500_returns = sp500_data.pct_change().dropna() * 100
gsci_returns = gsci_data.pct_change().dropna() * 100

# Combine the returns into a single DataFrame
combined_returns = pd.DataFrame({
    'SP500_Returns': sp500_returns,
    'GSCI_Returns': gsci_returns
})

# Calculate the annual mean return and standard deviation in percentage
annual_mean_return = combined_returns.resample('Y').mean()
annual_std_return = combined_returns.resample('Y').std()

# Calculate the total mean return and standard deviation for the entire period in percentage
total_mean_return = combined_returns.mean()
total_std_return = combined_returns.std()

# Calculate correlation of daily returns
correlation = combined_returns['SP500_Returns'].corr(combined_returns['GSCI_Returns'])

# Plotting
# Annual Mean Returns
annual_mean_return.plot(kind='bar', title='Annual Mean Returns')
plt.ylabel('Mean Return (%)')
plt.show()

# Annual Standard Deviations
annual_std_return.plot(kind='bar', title='Annual Standard Deviation of Returns')
plt.ylabel('Standard Deviation (%)')
plt.show()

# Scatter Plot for Correlation
plt.scatter(combined_returns['SP500_Returns'], combined_returns['GSCI_Returns'])
plt.title('Daily Returns: S&P 500 vs GSCI')
plt.xlabel('S&P 500 Returns (%)')
plt.ylabel('GSCI Returns (%)')
plt.show()

# Output the results
print("Total Mean Return (in %):\n", total_mean_return)
print("Total Standard Deviation of Returns (in %):\n", total_std_return)
print("Correlation between S&P 500 and GSCI Returns:", correlation)

plt.plot(combined_returns['GSCI_Returns']
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Define the tickers and the start date
sp500_ticker = "^GSPC"  # Ticker for S&P 500
gsci_ticker = "^SPGSCI"     # An ETF approximating GSCI
start_date = "2006-02-03"
dbc_ticker = "DBC"

# Fetch the historical data from Yahoo Finance
sp500_data = yf.download(sp500_ticker, start=start_date)['Adj Close']
dbc_data = yf.download(dbc_ticker, start=start_date)['Adj Close']
gsci_data = yf.download(gsci_ticker, start=start_date)['Adj Close']

# Calculate daily returns in percentage
sp500_returns = sp500_data.pct_change().dropna()
gsci_returns = gsci_data.pct_change().dropna()
dbc_returns = dbc_data.pct_change().dropna()

# Calculate cumulative returns
sp500_cumulative = ((1 + sp500_returns).cumprod() - 1)*100
gsci_cumulative = ((1 + gsci_returns).cumprod() - 1)*100
dbc_cumulative = ((1 + dbc_returns).cumprod() - 1)*100

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(sp500_cumulative, label='S&P 500 Cumulative Log Returns')
# plt.plot(gsci_cumulative, label='GSCI Cumulative Log Returns')
plt.plot(dbc_cumulative, label='DBC Cumulative Log Returns')
plt.title('Cumulative Log Returns of DBC and S&P 500')
plt.xlabel('Date')
plt.ylabel('Cumulative Log Return')
plt.legend()
plt.show()
```

```python
import pandas as pd
import yfinance as yf
from datetime import datetime

# Define the tickers
sp500_ticker = "^GSPC"  # Ticker for S&P 500
gsci_ticker = "DBC"     # An ETF approximating GSCI

# Define the start date and end date
start_date = "2020-02-01"
end_date = datetime.today().strftime('%Y-%m-%d')

# Fetch the historical data from Yahoo Finance
sp500_data = yf.download(sp500_ticker, start=start_date, end=end_date)['Adj Close']
gsci_data = yf.download(gsci_ticker, start=start_date, end=end_date)['Adj Close']

# Calculate daily returns
sp500_daily_returns = sp500_data.pct_change().dropna()
gsci_daily_returns = gsci_data.pct_change().dropna()

# Calculate the number of years in the period
number_of_years = (sp500_data.index[-1] - sp500_data.index[0]).days / 365.25

# Calculate the annualized return for SP500 and GSCI
sp500_annualized_return = (sp500_data[-1] / sp500_data[0]) ** (1 / number_of_years) - 1
gsci_annualized_return = (gsci_data[-1] / gsci_data[0]) ** (1 / number_of_years) - 1

# Calculate annual standard deviation
annual_trading_days = 252
sp500_annual_std = sp500_daily_returns.std() * (annual_trading_days ** 0.5)
gsci_annual_std = gsci_daily_returns.std() * (annual_trading_days ** 0.5)

# Output the results
print(f"Annualized Return for S&P 500 from 1991-05-01 to Present: {sp500_annualized_return * 100:.2f}%")
print(f"Annualized Return for DBC from 1991-05-01 to Present: {gsci_annualized_return * 100:.2f}%")
print(f"Annual Standard Deviation for S&P 500: {sp500_annual_std * 100:.2f}%")
print(f"Annual Standard Deviation for DBC: {gsci_annual_std * 100:.2f}%")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define the ticker symbols
tickers = ['DBC', 'SPY', 'TLT', 'IEF', 'SHY', 'XSVM', 'IGSB', '^SPGSCI', '^SBBMGLU']

# Fetch historical data
data = yf.download(tickers, start="2006-02-03", end="2023-03-31")['Adj Close']

# Calculate daily log returns
log_returns = np.log(data / data.shift(1))

# Calculate cumulative log returns
cumulative_log_returns = log_returns.cumsum()

# Plot cumulative log returns
plt.figure(figsize=(14, 7))
for ticker in cumulative_log_returns.columns:
    plt.plot(cumulative_log_returns.index, cumulative_log_returns[ticker], label=ticker)

plt.legend()
plt.title("Cumulative Log Return of Portfolio Assets")
plt.xlabel("Date")
plt.ylabel("Cumulative Log Returns")
plt.show()
```
