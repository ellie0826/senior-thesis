```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define the tickers for GSCI, DBC, S&P 500, and a proxy for US Treasury Bills (using a short-term treasury ETF, e.g., SHV)
tickers = {
    'GSCI': '^SPGSCI',  # Note: GSCI's direct ticker might not be available, so using S&P 500 as a placeholder
    'DBC': 'DBC',
    'SP500': '^GSPC',
    'USTBills': 'SHV' # Short-term treasury ETF as a proxy for US Treasury Bills
}

# Define the start and end dates for the historical data
start_date = '2008-09-01'
end_date = '2023-09-01'

# Fetch historical data
data = yf.download(list(tickers.values()), start=start_date, end=end_date)['Adj Close']

# Calculate log returns
log_returns = np.log(data / data.shift(1))

# Calculate cumulative log returns
cumulative_log_returns = log_returns.cumsum()

# Plot the cumulative log returns
cumulative_log_returns.plot(figsize=(14, 7))
plt.title('Cumulative Log Returns of GSCI, DBC, S&P 500, and US Treasury Bills')
plt.ylabel('Cumulative Log Returns')
plt.xlabel('Date')
plt.legend(tickers.keys())
plt.grid(True)
plt.show()
```
